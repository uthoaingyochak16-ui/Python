```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    "Name": ["Rahim", "Karim", "Ayesha", "Nabila"],
    "CGPA": [3.45, 3.20, 3.80, 3.60]
}

df = pd.DataFrame(data)

# Line plot
plt.figure()
plt.plot(df["Name"], df["CGPA"])
plt.title("CGPA Line Plot")
plt.xlabel("Students")
plt.ylabel("CGPA")
# plt.show()

# Bar plot
plt.figure()
plt.bar(df["Name"], df["CGPA"])
plt.title("CGPA Bar Plot")
plt.xlabel("Students")
plt.ylabel("CGPA")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create the dataset
data = {
    'surname': ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Rodriguez', 'Martinez'],
    'grade_point': [4.0, 3.0, 4.0, 2.0, 3.0, 4.0, 1.0, 3.0, 2.0, 4.0]
}
df = pd.DataFrame(data)

# 2. Create the graph
df.plot(kind='bar', x='surname', y='grade_point', color='lightgreen')

# 3. ADD GRID LINES HERE
# linestyle='--' makes them dashed, alpha=0.7 makes them slightly transparent
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 4. Labels and formatting
plt.title('Student Performance (with Grid Lines)')
plt.ylabel('Grade Point')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('lineplot.png',dpi=300,bbox_inches='tight')

plt.show()
```
