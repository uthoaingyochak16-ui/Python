```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correct way to load a seaborn dataset
sns.get_dataset_names()
```

```python
penguins=sns.load_dataset('penguins')
penguins.head(20)
```

```python
penguins[['species','island']].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
penguins.keys()
penguins.columns.to_list()
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,6))
sns.set_theme('notebook')
# sns.scatterplot(data=penguins, x='island', y='body_mass_g',hue='sex')
sns.stripplot(data=penguins, x='island', y='body_mass_g',hue='sex',dodge=True,jitter=True)
plt.xticks(rotation=45)
sns.despine(left=True)
```

```python
sns.set_context('notebook')
sns.histplot(data=penguins,x='body_mass_g',y='species',hue='sex')
```

```python
sns.jointplot(data=penguins,x="body_mass_g",y='flipper_length_mm',kind='reg')
```

```python
sns.pairplot(data=penguins,hue='sex',diag_kind='kde')
```
