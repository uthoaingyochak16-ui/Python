```python
import json

def search_json_file(file_path, query):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        query = query.strip().lower()
        results = []

        for item in data:
            # City, Address ebong Name er moddhe query khujbe
            city = str(item.get('city', '')).lower()
            address = str(item.get('address', '')).lower()
            name = str(item.get('name', '')).lower()

            if query in city or query in address or query in name:
                results.append(item)

        if results:
            print(f"\n--- {len(results)} ti Result Pauya Geche ---")
            for i, res in enumerate(results, 1):
                print(f"{i}. Name: {res['name']}")
                print(f"   Address: {res['address']}")
                print(f"   Phone: {res['phone1']}")
                print("-" * 30)
        else:
            print("Kono result pauya jayni.")

    except Exception as e:
        print(f"Error: {e}")

# Use it like this:
file_name = 'ContactData.json'
search_term = input("Search Location (e.g., Dhaka, Gazipur, Shantinagar): ")
search_json_file(file_name, search_term)
```

```python
import json

def show_all_office_details():
    try:
        # JSON file open kora
        with open('ContactData.json', 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        print("--- Office Directory Search (Full Details) ---")
        query = input("Location ba Office er nam likhun: ").strip().lower()
        
        if not query:
            print("Kichu ekta likhe search korun.")
            return

        found = False
        for office in data:
            # Search fields
            name = str(office.get('name', '')).lower()
            address = str(office.get('address', '')).lower()
            city = str(office.get('city', '')).lower()

            if query in name or query in address or query in city:
                found = True
                print("\n" + "█" * 60) # Divider
                print(f" OFFICE: {office.get('name', '').upper()}")
                print("█" * 60)
                
                # Office er shob details show kora hocche
                for key, value in office.items():
                    # SI (sl) skip kora hocche
                    if key.lower() == 'sl':
                        continue
                    
                    # Key name format kora
                    display_key = key.replace('_', ' ').capitalize()
                    
                    # Jodi value thake, tobe seta purno length-e show korbe
                    if value is not None and value != "":
                        # Ekhane kono width limit rakha hoyni, tai data hide hobe na
                        print(f"{display_key}: {value}")
                
                print("-" * 60)

        if not found:
            print(f"\n'{query}' diye kono office pauya jayni.")

    except FileNotFoundError:
        print("Error: 'ContactData.json' file-ti same folder-e rakhenni.")
    except Exception as e:
        print(f"Error hoyeche: {e}")

if __name__ == "__main__":
    show_all_office_details()
```

```python
import json

def show_office():
    try:
        with open('ContactData.json', 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        query = input("Search (Name/City/Address): ").strip().lower()
        if not query: return

        for o in data:
            # সার্চ করার জন্য সব ভ্যালুকে একসাথে করা
            content = " ".join([str(v) for v in o.values()]).lower()
            
            if query in content:
                print(f"\n{'='*50}")
                print(f"{o.get('name', 'N/A').upper()}".center(50))
                print(f"{'-'*30}".center(50))
                
                # ঠিকানা
                print(f"{o.get('address', '')}".center(50))
                if o.get('city'): print(f"{o.get('city', '')}".center(50))
                print() 

                # ডাইনামিক কন্টাক্ট চেকার (Space বা Capital Letter যাই থাকুক কাজ করবে)
                for key, value in o.items():
                    if not value or value == "": continue
                    
                    k_lower = key.lower()
                    if 'phone' in k_lower:
                        print(f"  📞 {value}")
                    elif 'mobile' in k_lower:
                        print(f"  📱 {value}")
                    elif 'email' in k_lower:
                        print(f"  ✉️ {value}")
                
                print(f"{'='*50}")

    except FileNotFoundError:
        print("Error: 'ContactData.json' ফাইলটি পাওয়া যায়নি।")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    show_office()
```

```python
require("dotenv").config();
const express = require("express");
const axios = require("axios");
const fs = require("fs");
const path = require("path");

const app = express();
app.use(express.json());

const { 
    GEMINI_API_KEY, PAGE_ACCESS_TOKEN, 
    VERIFY_TOKEN, APPS_SCRIPT_URL 
} = process.env;


const officeData = JSON.parse(fs.readFileSync(path.join(__dirname, "ContactData.json"), "utf8"));


function formatOfficeInfo(location) {
    if (!location) return "";
    const query = location.toLowerCase();
    
    const matched = officeData.filter(o => {
        const content = Object.values(o).join(" ").toLowerCase();
        return content.includes(query);
    });

    if (matched.length === 0) return "দুঃখিত, আপনার দেওয়া লোকেশনে আমাদের কোনো অফিস পাওয়া যায়নি।";

    let output = "আপনার সুবিধার জন্য আমাদের নিকটস্থ অফিসের বিস্তারিত তথ্য নিচে দেওয়া হলো:\n";
    
    matched.forEach(o => {
        output += `\n==================================================\n`;
        output += `${o.name ? o.name.toUpperCase() : 'N/A'}\n`;
        output += `------------------------------\n`;
        output += `${o.address || ''}\n${o.city || ''}\n\n`;

        // ডাইনামিক কন্টাক্ট চেকার (Phone, Mobile, Email)
        Object.entries(o).forEach(([key, value]) => {
            if (!value) return;
            const k = key.toLowerCase();
            if (k.includes('phone')) output += `  📞 ${value}\n`;
            else if (k.includes('mobile')) output += `  📱 ${value}\n`;
            else if (k.includes('email')) output += `  ✉️ ${value}\n`;
        });
        output += `==================================================\n`;
    });

    return output;
}

// ৩. Typing Indicator এবং মেসেজ পাঠানোর ফাংশন
async function sendFBResponse(id, text) {
    try {
        await axios.post(`https://graph.facebook.com/v19.0/me/messages?access_token=${PAGE_ACCESS_TOKEN}`, {
            recipient: { id },
            sender_action: "typing_on"
        });

        const typingDelay = Math.min(text.length * 20, 3000); // টেক্সট অনুযায়ী টাইপিং ডিলে
        await new Promise(resolve => setTimeout(resolve, typingDelay));

        await axios.post(`https://graph.facebook.com/v19.0/me/messages?access_token=${PAGE_ACCESS_TOKEN}`, {
            recipient: { id },
            message: { text }
        });
    } catch (e) { console.error("FB Send Error"); }
}

// ৪. AI ইঞ্জিন (শুধুমাত্র Gemini 3.1 Flash Lite)
async function askAI(userMsg) {
    const prompt = `Extract JSON: {"intent": "info_sharing"|"qna", "data": {"name":"..","phone":"..","problem":"..","location":".."}, "answer": "Bengali response"}. User: ${userMsg}`;
    
    try {
        const url = `https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key=${GEMINI_API_KEY}`;
        const res = await axios.post(url, { contents: [{ parts: [{ text: prompt }] }] }, { timeout: 7000 });
        const raw = res.data.candidates[0].content.parts[0].text;
        const match = raw.match(/\{[\s\S]*\}/);
        return match ? JSON.parse(match[0]) : null;
    } catch (e) {
        console.error("AI Error");
        return null;
    }
}

// ৫. মেইন ওয়েবহুক হ্যান্ডলার (Scalable)
app.post("/webhook", async (req, res) => {
    res.status(200).send("EVENT_RECEIVED");
    const entry = req.body.entry?.[0]?.messaging?.[0];
    if (!entry?.message?.text) return;

    const senderId = entry.sender.id;
    const userMsg = entry.message.text;

    // AI এনালাইসিস (Async)
    const ai = await askAI(userMsg);
    if (!ai) return;

    if (ai.intent === "info_sharing") {
        const { name, phone, location, problem } = ai.data;

        // আপনার দেওয়া কনভারসেশন ফ্লো (শুদ্ধ বাংলা বানানে)
        if (!name || name === "..") {
            return sendFBResponse(senderId, "আচ্ছা, আপনার সম্পূর্ণ নাম লিখুন।");
        }
        if (!phone || phone === "..") {
            return sendFBResponse(senderId, `ধন্যবাদ ${name}! এখন আপনার মোবাইল নম্বরটি লিখুন।`);
        }
        if (!problem || problem === "..") {
            return sendFBResponse(senderId, "এখন আপনার সমস্যাটি সংক্ষিপ্তভাবে লিখুন।");
        }
        if (!location || location === "..") {
            return sendFBResponse(senderId, "এখন আপনার বর্তমান ঠিকানাটি লিখুন। এতে আমি আপনাকে আমাদের নিকটতম অফিস/সেল/সেন্টার-এর ঠিকানা প্রদান করতে পারব।");
        }

        // শিটে ডাটা সেভ (Background)
        axios.post(APPS_SCRIPT_URL, { 
            action: "appendRow", 
            sheetName: "UserData", 
            rowData: [name, phone, location, problem, new Date().toLocaleString()] 
        }).catch(() => {});

        const officeInfo = formatOfficeInfo(location);
        const finalReply = `ধন্যবাদ ${name}! আমাদের মূল্যবান সময় দেওয়ার জন্য। আপনার দেওয়া তথ্যগুলো আমাদের ডাটাবেসে সফলভাবে সংরক্ষিত হয়েছে। কিছুক্ষণের মধ্যেই আমাদের একজন প্রতিনিধি আপনার সাথে যোগাযোগ করবেন।\n\n${officeInfo}`;
        
        return sendFBResponse(senderId, finalReply);
    }

    sendFBResponse(senderId, ai.answer);
});

app.get("/webhook", (req, res) => {
    if (req.query["hub.verify_token"] === VERIFY_TOKEN) res.send(req.query["hub.challenge"]);
    else res.sendStatus(403);
});

app.listen(process.env.PORT || 10000, () => console.log("🚀 Quantum Lightweight Bot is Live!"));
```
