### why we need operator function

```python
import operator

list1 = [1, 2, 3]
list2 = [4, 5, 6]

# এখানে operator.add একটি ফাংশন হিসেবে কাজ করছে
result = list(map(operator.add, list1, list2))

print(result) 
# আউটপুট: [5, 7, 9] (অর্থাৎ: 1+4, 2+5, 3+6)
```

#### If you need to find the result of integer division and modulus, you can use the divmod function as a shortcut:


```python
quotient, remainder = divmod(9, 4)
print(quotient,remainder)
```
