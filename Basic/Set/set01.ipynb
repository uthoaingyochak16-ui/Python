```python
## Interscetion

x={1, 2, 3, 4, 5}.intersection({3, 4, 5, 6}) # {3, 4, 5}
y={1, 2, 3, 4, 5} & {3, 4, 5, 6}  # {3, 4, 5}

print(x)
print(y)
```

### Union set

```python
x={1, 2, 3, 4, 5}.union({3, 4, 5, 6}) # {1, 2, 3, 4, 5, 6}
y={1, 2, 3, 4, 5} | {3, 4, 5, 6} # {1, 2, 3, 4, 5, 6}

print(x)
print(y)
```

```python
# Difference
x={1, 2, 3, 4}.difference({2, 3, 5}) # {1, 4}
y={1, 2, 3, 4} - {2, 3, 5} # {1, 4}

print(x)
print(y)
```

```python
# Symmetric difference with

# this is alternative of differnce set
{1, 2, 3, 4}.symmetric_difference({2, 3, 5}) # {1, 4, 5}
{1, 2, 3, 4} ^ {2, 3, 5} # {1, 4, 5}
```

```python
# Superset check
{1, 2}.issuperset({1, 2, 3}) # False
{1, 2} >= {1, 2, 3} # False
```

```python
# Subset check
{1, 2}.issubset({1, 2, 3}) # True
{1, 2} <= {1, 2, 3} # True
```

```python
# Disjoint check
{1, 2}.isdisjoint({3, 4}) # True
{1, 2}.isdisjoint({1, 4}) # False
```

```python
# Existence check
2 in {1,2,3} # True
4 in {1,2,3} # False
4 not in {1,2,3} # True
```

```python
## with single elements

s = {1,2,3}
s.add(4)
s.discard(3)
s.discard(5)
s.remove(2)
print(s) 
```

```python
## difference between update and  add

s = {1, 2, 3}
s.update([4, 5, 6]) 
s.update({20,21})
s.update((50,80))
s.add(31)
print(s) # আউটপুট: {1, 2, 3, 4, 5, 6}
```
