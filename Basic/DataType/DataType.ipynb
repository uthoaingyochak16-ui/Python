```python
a=set('abdracadabra')
print(a)
```

```python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(basket) 
```

#####  Frozen Sets - They are immutable and new elements cannot added after its defined.


```python
b = frozenset('asdfagsa')
print(b)
```

```python
cities = frozenset(["Frankfurt", "Basel","Freiburg"])
print(cities)
```
