```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = [1, 2, 3, 4, 5, 6]
S25 = [120, 200, 300, 189, 178, 255]
S24 = [200, 230, 190, 260, 288, 299]

# 1. Setup the positions
x = np.arange(len(months))  # The label locations [0, 1, 2, 3, 4, 5]
width = 0.35                # The width of the bars
fig=plt.figure(num="A unique title.",facecolor='skyblue')
fig, ax = plt.subplots(figsize=(10, 6),facecolor='skyblue',edgecolor='red')

# 2. Create bars side-by-side
# Subtract half width from x for the first set, add half for the second
rects1 = ax.bar(x - width/2, S25, width, label='2025 Sales', color='#2ca02c') # Green
rects2 = ax.bar(x + width/2, S24, width, label='2024 Sales', color='#1f77b4') # Blue

# 3. Add details for "Perfection"
ax.set_xlabel('Month')
ax.set_ylabel('Sales Volume')
ax.set_title('Comparison of Sales: 2024 vs 2025')
ax.set_xticks(x)
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
ax.legend()

# 4. Add data labels on top of bars
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0,0].plot(x, np.sin(x))
ax[0,0].set_title("Sin")

ax[0,1].plot(x, np.cos(x))
ax[0,1].set_title("Cos")

ax[1,0].plot(x, np.tan(x))
ax[1,0].set_title("Tan")

ax[1,1].plot(x, x**2)
ax[1,1].set_title("Square")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales_2025 = [120, 200, 300, 189, 178, 255]
sales_2024 = [200, 230, 190, 260, 288, 299]

# Bar position
x = np.arange(len(months))
width = 0.35

# Create graph
plt.figure(figsize=(10,6))

# Bars
plt.bar(x - width/2, sales_2025, width, label='2025 Sales', color='green')
plt.bar(x + width/2, sales_2024, width, label='2024 Sales', color='blue')

# Labels & title
plt.title("Comparison of Sales: 2024 vs 2025")
plt.xlabel("Month")
plt.ylabel("Sales Volume")

# X-axis names
plt.xticks(x, months)

# Legend
plt.legend()

# Show graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
x = np.arange(len(months))

# 4 datasets
data = [
    [120, 200, 300, 189, 178, 255],
    [200, 230, 190, 260, 288, 299],
    [150, 180, 220, 210, 230, 250],
    [180, 210, 240, 200, 220, 260]
]

labels = ['Disguise 1', 'Disguise 2', 'Disguise 3', 'Disguise 4']
colors = ['#2ca02c', '#1f77b4', '#ff7f0e', '#d62728']

width = 0.2

plt.figure(figsize=(10,6))

# Loop to plot each dataset
for i in range(len(data)):
    plt.bar(x + (i - 1.5)*width, data[i], width=width, label=labels[i], color=colors[i])

plt.xticks(x, months)
plt.ylabel('Values')
plt.title('Monthly Data for 4 Disguise Bars')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create 100 random data points for 5 different variables
n = 100
dim1_x = np.random.rand(n)
dim2_y = np.random.rand(n)
dim3_z = np.random.rand(n)
dim4_color = np.random.rand(n)  # Values for color mapping
dim5_size = np.random.rand(n) * 500  # Values for bubble size

# 2. Setup the 3D Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 3. Plot using scatter
# c=dim4_color sets the 4th dimension
# s=dim5_size sets the 5th dimension
img = ax.scatter(dim1_x, dim2_y, dim3_z, 
                 c=dim4_color, 
                 s=dim5_size, 
                 alpha=0.6, 
                 cmap='viridis')

# 4. Add a Colorbar to explain the 4th dimension
fig.colorbar(img, ax=ax, label='Dimension 4 (Color)')

# 5. Add Labels
ax.set_xlabel('Dim 1 (X)')
ax.set_ylabel('Dim 2 (Y)')
ax.set_zlabel('Dim 3 (Z)')
plt.title('5D Data Visualization')

plt.show()
```
