##   Bitwise  Operation

```python
import cv2
import numpy as np

img1 = cv2.imread('a.png')
img2 = cv2.imread('b.png')

dest1 = cv2.bitwise_and(img2, img1, mask = None)
dest2 = cv2.bitwise_or(img2, img1, mask = None)
dest3 = cv2.bitwise_xor(img1, img2, mask = None)
cv2.imshow('A', img1)
cv2.imshow('B', img2)
cv2.imshow('AND', dest1)
cv2.imshow('OR', dest2)
cv2.imshow('XOR', dest3)
cv2.imshow('NOT A', cv2.bitwise_not(img1))
cv2.imshow('NOT B', cv2.bitwise_not(img2))
if cv2.waitKey(0) & 0xff == 27:
 cv2.destroyAllWindows()
```

## Handling Mouse Events

```python
import cv2 as cv
# mouse callback function
def drawfunction(event,x,y,flags,param):
    if event == cv.EVENT_LBUTTONDBLCLK:
        cv.circle(img,(x,y),20,(255,255,255),-1)
img = cv.imread('gray.jpeg')
cv.namedWindow('image')
cv.setMouseCallback('image',drawfunction)
while(1):
 cv.imshow('image',img)
 key=cv.waitKey(1)
 if key == 27:
    break
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
# mouse callback function
drawing=True
shape='r'

def draw_circle(event,x,y,flags,param):
 global x1,x2
 if event == cv.EVENT_LBUTTONDOWN:
    drawing = True
    x1,x2 = x,y
 elif event == cv.EVENT_LBUTTONUP:
    drawing = False
    if shape == 'r':
        cv.rectangle(img,(x1,x2),(x,y),(0,255,0),-1)
    if shape == 'l':
        cv.line(img,(x1,x2),(x,y),(255,255,255),3)
    if shape=='c':
        cv.circle(img,(x,y), 10, (255,255,0), -1)
img = cv.imread('gray.jpeg')
cv.namedWindow('image')
cv.setMouseCallback('image',draw_circle)
while(1):
 cv.imshow('image',img)
 key=cv.waitKey(1)
 if key==ord('1'):
    shape='r'
 if key==ord('2'):
    shape='l'
 if key==ord('3'):
    shape='c'
 #print (shape)
 if key == 27:
    break
cv.destroyAllWindows()
```

## Capture Video from Camera

```python
import cv2 as cv

# Camera start
cam = cv.VideoCapture(0)

# Video writer setup
cc = cv.VideoWriter_fourcc(*'XVID')
# Frame size camera-r resolution-er shathe mil thakte hobe
file = cv.VideoWriter('output.avi', cc, 20.0, (480, 380))

if not cam.isOpened():
    print("Error opening camera")
    exit()

while True:
    ret, frame = cam.read()
    
    if not ret:
        print("Error in retrieving frame")
        break

    # Frame-ti file-e save korchi (BGR format-ei)
    file.write(frame)

    # Video display korchi (OpenCV window-te)
    cv.imshow('Live Video', frame)

    # 'q' press korle bondho hobe
    if cv.waitKey(1) == ord('q'):
        break

# Shob kichu release kora
cam.release()
file.release()
cv.destroyAllWindows()
for i in range(1, 5):
    cv.waitKey(1)
```

```python
import cv2 as cv

cam = cv.VideoCapture(0)

if not cam.isOpened():
    print("Error: Could not open camera.")
    exit()

# Get resolution
frame_width = int(cam.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(cam.get(cv.CAP_PROP_FRAME_HEIGHT))
size = (frame_width, frame_height)

# --- THE FIX IS HERE ---
# Use 'mp4v' and change the extension to '.mp4'
fourcc = cv.VideoWriter_fourcc(*'mp4v') 
out = cv.VideoWriter('output.mp4', fourcc, 20.0, size)
# -----------------------

print("Recording... Press 'q' to stop.")

while True:
    ret, frame = cam.read()
    if not ret:
        break

    out.write(frame)
    cv.imshow('Live Video', frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
out.release()
cv.destroyAllWindows()
```
